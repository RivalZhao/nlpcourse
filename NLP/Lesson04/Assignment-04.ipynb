```python
from pathlib import Path
from tqdm import tqdm
import jieba
import pickle
import re
import logging
import gensim
from gensim.models import Word2Vec
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import matplotlib
```

```python
def testWord2Vec():
    #model = gensim.models.Word2Vec.load('./model/wiki_news_all_500')
    model = gensim.models.Word2Vec.load('E:/CloudStationBackup/PyCharm/NLP-ClassFile/nlpcourse/NLP/Lesson04/model/wiki_news_all_500')
    print(model.similarity('图书', '书籍'))
    print(model.most_similar(['美丽']))
    tsne_plot(model)


def tsne_plot(model):
    %matplotlib inline
    "Creates and TSNE model and plots it"
    labels = []
    tokens = []
    matplotlib.rcParams['font.sans-serif'] = ['SimHei']
    matplotlib.rcParams['font.family'] = 'sans-serif'

    for word in model.wv.vocab:
        tokens.append(model[word])
        labels.append(word)

    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
    new_values = tsne_model.fit_transform(tokens)

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])

    plt.figure(figsize=(16, 16))
    for i in range(len(x)):
        plt.scatter(x[i], y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')
    #plt.show()


```

```python
testWord2Vec()
```
